## 'read_csv_rows'

- Function name: 'read_csv_rows'
- parameter: 'str' path to csv file
- Return type: 'list[dict[str,str]]

This function reads csv files and gives me a data structure to work with

```python
def read_csv_rows(filename:str) -> list[dict[str,str]]:
    """Read the rows of a CSV"""
    result: list[dict[str,str]] = []
    # Open the file to read it
    file_handle = open(filename, "r")

    #Make a dictreader
    csv_reader = DictReader(file_handle)
    # copy each row of csvreader into our list
    for row in csv_reader:
        result.append(row)
    #close the file
    file_handle.close()

    return result
    
```
